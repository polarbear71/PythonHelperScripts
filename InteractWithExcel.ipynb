```python
import pandas as pd

file = "example.xlsx"

# load spreadsheet
x1 = pd.ExcelFile(file)

# Print sheet names
print(x1.sheet_names)

# Load a sheet into a DataFrame by name: df1
df1 = x1.parse("Sheet1")

print(df1)
```

```python
df1 = df1.append({"Vorname" : "Karl", "Nachname" : "Eschenbach"}, ignore_index=True)
df1
```

```python
df1.to_excel('output.xlsx')
```

```python
df2 = pd.DataFrame(
    {'Vorname' : ['Albert', 'Mario', 'Norbert'],
    'Nachname' : ['Carlsson', 'Wiedenfall', 'Notnagel']},
    index = [0, 1, 2])
```

```python
df2
```

```python
df3 = pd.concat([df1, df2])
df3
```

```python
with pd.ExcelWriter ('output.xlsx') as writer:
    df1.to_excel(writer, sheet_name='Sheet1')
    df2.to_excel(writer, sheet_name='Sheet2')
    df3.to_excel(writer, sheet_name='Sheet3')
```
