```python
import pandas as pd
from random import *
```

```python
list = [i*randint(0, 10) for i in range(0,100)]

# Generate a dataframe with numbers
df = pd.DataFrame({'x' : [i for i in range(0,100)],
                  'y' : list,
                  'z' : [i*2 for i in range(0,100)]},
                 index = [i for i in range(0,100)])
```

```python
print(df.plot.scatter(x='x', y='y'))
```

```python
print(df.plot.hist(bins=8, alpha = 0.5))
```
